##Matplot Lb Workflow

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1. Preapare the data

x=[1,2,3,4,5]
y=[11,22,33,44,55]


#2. Setup the plot

fig,ax= plt.subplots(figsize=(10,10))

#3. Plot data
ax.plot(x,y)

#4. Customize plot

ax.set(title="Simple Plot",
       xlabel="x-Axis",
       ylabel="y-Axis")
plt.show()
```

```python
#Make some data
x=np.random.randn(1000)

fig, ax= plt.subplots()
ax.hist(x);
```

```python
#Subplot option-1

fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax1.plot(x,x/2);
ax2.hist(x);
ax3.scatter(x,np.sin(x));
ax4.boxplot(x);
```

## Plotting from pandas as pd

```python
import pandas as pd

car_sales=pd.read_csv('https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/car-sales.csv')
```

```python
car_sales
```

```python
car_sales['Price']=car_sales['Price'].str.replace('[\$\,\.]','')
car_sales
```

```python
car_sales['Price']=car_sales["Price"].str[:-2]
```

```python
car_sales
```

```python
car_sales["Sale_Date"]=pd.date_range("1/1/2020", periods=len(car_sales))
```

```python
car_sales
```

```python
car_sales['Total_sales']=car_sales['Price'].cumsum()
```

```python
car_sales
```

```python
car_sales['Total_sales']=car_sales['Price'].astype(int).cumsum()
car_sales
```

```python
#Plotting the data
car_sales.plot(x='Sale_Date',y='Total_sales')
```

```python
car_sales['Price']=car_sales['Price'].astype(int)
car_sales.plot(x="Odometer (KM)",y='Price', kind='scatter')
```

```python
car_sales.plot(x='Make',y='Odometer (KM)',kind='bar')
```

```python
car_sales['Odometer (KM)'].plot.hist()
```

```python
car_sales['Odometer (KM)'].plot.hist(bins=20)
```

```python
car_sales['Odometer (KM)'].plot(kind='kde');
```

## Lets try another dataset

```python
heart_dis=pd.read_csv('https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/heart-disease.csv')
```

```python
heart_dis
```

```python
heart_dis['age'].plot(kind='kde');
```

```python
heart_dis['age'].plot.hist(bins=10);
```

```python
heart_dis.plot.hist(subplots=True,figsize=(10,30));
```

```python
over_50=heart_dis[heart_dis['age']>50]
over_50.head()
```

```python
# Pyplot Method
over_50.plot(kind='scatter',x='age',y='chol',c='target')
```

```python
# Plotting using OO Method
fig,ax=plt.subplots(figsize=(10,6));
# ax.plot(x=over_50['age'],y=over_50['chol'],c=over_50['target'])
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax)
# ax.set_xlim([45,100])
```

```python
fig,ax= plt.subplots(figsize=(10,6))

#Plot the data 
scatter= ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target']);

# Customize the plot
ax.set(title="Heart Diseaes and Cholesoral Levels",
       xlabel='Age',
       ylabel='Cholestro')

# Add legend
ax.legend(*scatter.legend_elements(),title='Target')


# Add a horizontal line

ax.axhline(over_50['chol'].mean(), linestyle='--')
```
